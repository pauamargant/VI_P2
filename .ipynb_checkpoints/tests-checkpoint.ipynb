```python
%load_ext autoreload 
%autoreload 2

from p1_graphs import *
import altair as alt
data = get_accident_data(fname="dataset_v1.csv",sample=True)
```

```python
select_bar = alt.selection_point(encodings=['x'], empty='all')
# make a bar chart of number of accidents by month, use altair transform_timeunit to extract month from date
chart1=alt.Chart(data).mark_point().encode(
    x=alt.X('month(date):T', title='Month'),
    y=alt.Y('count()', title='Number of Accidents'),
    color=alt.condition(select_bar, 'Severity', alt.value('lightgray'))
).add_params(select_bar)

# make scatter plot of accidents 
```

```python
# make scatter plot of accidents by month
chart2=alt.Chart(data).mark_circle().encode(
    x=alt.X('month(date):T', title='Month'),
    y=alt.Y('count()', title='Number of Accidents'),
    color=alt.condition(select_bar, alt.value('red'), alt.value('lightgray'))
).add_selection(select_bar).transform_filter(select_bar)
chart3 = alt.Chart(data).mark_point().encode(
    x=alt.X('month(date):T', title='Month'),
    y=alt.Y('count()', title='Number of Accidents'),
    color=alt.condition(select_bar, alt.value("green"), alt.value('lightgray'))
).add_params(select_bar)

chart1 | chart2 | chart3
```

```python
mapa = get_map()
```

```python
# dro pna in lat and long
df = data.dropna(subset=["LATITUDE", "LONGITUDE"])
print(df.shape)
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(df.LONGITUDE, df.LATITUDE)
)
gdf = gdf.set_crs(epsg=4326, inplace=True).to_crs("ESRI:102003")
print(gdf.shape)
gdf = gpd.sjoin(gdf, mapa.reset_index(), how="left", op="intersects")
print(gdf.shape)

gdf
```

```python
mapa["id"]=mapa.index
```

```python
alt.Chart(mapa)\
    .mark_geoshape()\
    .encode(
        color=alt.Color(
            "count()",
            title="Number of accidents",
            scale=alt.Scale(scheme="greenblue"),
        ),
        tooltip=["h3_polyfill:N", "counts:Q"],
    )\
    .transform_lookup(
        lookup="h3_polyfill",
        from_=alt.LookupData(gdf, "h3_polyfill",["h3_polyfill"]),
    )\
    .project(type="identity", reflectY=True)
```

```python
gdf_subset = gdf[["h3_polyfill"]]
gdf_subset.columns=["id"]
print(gdf_subset.columns)
print(mapa.columns)
```

```python
alt.Chart(mapa)\
.mark_geoshape()\
.transform_aggregate(
    counts='count()',
    groupby=['id']
)\
.encode(
    color=alt.Color(
        "counts:Q",
        title="Number of accidents",
        scale=alt.Scale(scheme="greenblue"),
    ),
    tooltip=["id:N", "counts:Q"],
)\
.transform_lookup(
    lookup="id",
    from_=alt.LookupData(gdf_subset, "id", ["id"]),
)\
.project(type="identity", reflectY=True)
```
